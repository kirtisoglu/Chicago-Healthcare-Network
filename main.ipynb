```python
geo_data = chicago.copy()
geo_data["color"] = (geo_data["population"] > 0).astype(int)


# Make sure CRS is lat/lon (EPSG:4326) for Plotly
geo_data = geo_data.to_crs(epsg=4326)

# Convert to GeoJSON
geojson_data = geo_data.__geo_interface__

# Plot with Plotly
fig = px.choropleth_mapbox(
    geo_data,
    geojson=geojson_data,
    locations=geo_data.index,        # index links rows to polygons
    color="color",                   # 0 or 1
    hover_data=["population"],       # show population on hover
    mapbox_style="carto-positron",
    center={"lat": 41.8781, "lon": -87.6298},  # Chicago
    zoom=9,
    opacity=0.6,
    width=900,   
    height=900,
    color_continuous_scale=["red", "yellow"]    # black=0, red=1
)

fig.update_layout(
    title="Chicago Census Blocks (Population vs. Empty)",
    margin={"r":0,"t":40,"l":0,"b":0}
)

fig.show()
```
